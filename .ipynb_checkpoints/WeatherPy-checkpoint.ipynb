```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import datetime as dt

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []
countries = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # Also grab the country of the city (if possible) 
    country = citipy.nearest_city(lat_lng[0], lat_lng[1]).country_code
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        countries.append(country)

# Print the city count to confirm sufficient count
print(len(cities))
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
print("Beginning Data Retrieval")
print("---------------------------------")

# Initialize record and set numbers
record_num = 1
set_num = 1

# Initiliaze other lists for the dataframe
found_cities = []
found_countries = []
clouds = []
dates = []
humidities = []
latitudes = []
longitudes = []
max_temps = []
wind_speeds = []
    
for x in range(len(cities)):
    # Gather the necessary information for the api query
    city_name = cities[x]
    country_code = countries[x]
    base_url = "http://api.openweathermap.org/data/2.5/weather?"
    query_url = base_url + "appid=" + api_key + f"&q={city_name},{country_code}&units=imperial"
    weather_json = requests.get(query_url).json()
     
    try:
        # Cloudiness
        clouds.append(weather_json["clouds"]["all"])
        # Date
        dates.append(weather_json["dt"])
        # Humidity
        humidities.append(weather_json["main"]["humidity"])
        # Latitude
        latitudes.append(weather_json["coord"]["lat"])
        # Longitude
        longitudes.append(weather_json["coord"]["lon"])
        # Max Temp
        max_temps.append(weather_json["main"]["temp_max"])
        # Wind Speed
        wind_speeds.append(weather_json["wind"]["speed"])
        # Found cities
        found_cities.append(city_name)
        # Found countries
        found_countries.append(country_code)
        
        #print(f"Processing {city_name} at {index}")
        print(f"Processing record {record_num} of set {set_num} | {city_name}")
        
        # Record and set numbers
        # If the record number is 50, reset the values
        if record_num == 50:
            record_num = 0
            # Increment the set every 50 records
            set_num += 1 
        # Increment the record every time the for loop runs
        record_num +=1
        
    except KeyError:
        print("City not found. Skipping...")
        
    
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Create new weather dataframe
city_weather = pd.DataFrame({"City": found_cities,
                            "Country": found_countries,
                             "Cloudiness": clouds,
                            "Date": dates,
                            "Humidity": humidities,
                            "Latitude": latitudes,
                            "Longitude": longitudes,
                            "Max Temp": max_temps,
                            "Wind Speed": wind_speeds
                            })

# Format the dataframe
# Change the max temp from K to 
# Capitalize country
city_weather["Country"] = city_weather["Country"].str.upper()
# Captialize city
city_weather["City"] = city_weather["City"].str.capitalize()

city_weather.head()
```

```python
# Export to a csv file
city_weather.to_csv(output_data_file, index=False, header=True)
```

### Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.

#### Latitude vs. Temperature Plot

```python
# Create plot
plt.scatter(city_weather["Latitude"], city_weather["Max Temp"], marker='o', facecolor='cornflowerblue', edgecolor='k',
            alpha=0.85, s=30)
today = dt.datetime.today().strftime("%m/%d/%Y")
plt.title(f"Latitude vs Max Temperature ({today})")
plt.xlabel("Latitude")
plt.ylabel("Max Temp (F)")
plt.show()

# Save as png
plt.savefig("output_data/max_temp.png")
```
